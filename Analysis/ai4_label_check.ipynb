```python
results = []

label_path = '/data/Aldhani/eoagritwin/fields/ai4boundaries/sentinel2/masks/'
countries = dirfinder(label_path)

for country in countries:
    print(country)
    c_files = getFilelist(f'{label_path}{country}', '.tif')
    for c_file in c_files:
        ds = gdal.Open(c_file)
        b1 = ds.GetRasterBand(1).ReadAsArray()
        area_b1 = np.sum(b1)
        if area_b1 == 0:
            continue
        else:
            b2 = ds.GetRasterBand(2).ReadAsArray()
            area_b2 = np.sum(b2)

            results.append({
                        'country': country,
                        'b1_pixel': area_b1,
                        'b2_pixel': area_b2,
                        'overlap': np.sum((b1 == 1) & (b2 ==1))
                    })

# convert and export
df = pd.DataFrame(results)
df.to_csv(f'ai4_label_comparison.csv', index=False)
```
