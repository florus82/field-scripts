```python
import sys
sys.path.append('/home/potzschf/repos/')
from helperToolz.helpsters import *
from helperToolz.evapo import *
import geopandas as gpd
```

```python
#### get path and rows of scenes that have data for the chosen AOI (e.g. Brandenburg)

# load shapefiles and 
ger = gpd.read_file(f'/data/Aldhani/eoagritwin/misc/gadm41_DEU_shp/gadm41_DEU_1.shp')

force_grid = gpd.read_file(f'/data/Aldhani/eoagritwin/misc/FORCE_footprint/FORCE_GRID_4326.shp')

for state in ger['NAME_1']:
    aoi = ger[ger['NAME_1'] == state]
    # check projections
    if aoi.crs != force_grid.crs:
        aoi = aoi.to_crs(force_grid.crs)

    force_touching = gpd.sjoin(
        force_grid,
        aoi,
        how='inner',
        predicate='intersects'  # or 'touches', depending on your need
    )

    out_str_x = "%XX%: " + " ".join(map(str, force_touching['Tile_X']))
    out_str_y = "%YY%: " + " ".join(map(str, force_touching['Tile_Y']))

    with open(f'/data/Aldhani/eoagritwin/force/tilefile/tiles_{state}.txt', 'w') as f:
        f.write(out_str_x + '\n')
        f.write(out_str_y)
```
